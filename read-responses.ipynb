```python
import numpy as np
import pandas as pd
import ast
import os
```

```python
results = pd.read_csv("responses/Ádám+survey+try_April+16,+2021_00.47.csv")
```

```python
results = results.T.assign(
    ImportId=lambda x: x[1].apply(ast.literal_eval).apply(lambda x: x["ImportId"],
                                                         )
)
```

```python
question_id = results.loc[
    results["ImportId"].apply(lambda x: 1 if "title" in x else None).dropna().index
].filter([5]).reset_index().rename(columns={5:"title","index":"Id"})
```

```python
answers = (
    results.loc[
        results[0].apply(lambda x: 1 if "${e://Field/" in x else None).dropna().index
    ]
    .drop(columns=["ImportId"])
    .assign(
        Id=lambda x: x[0]
        .str.split("}")
        .apply(lambda x: x[0])
        .str.split("/")
        .apply(lambda x: x[-1])
    ).iloc[:,2:].reset_index(drop=True)
)
```

```python
results = pd.merge(question_id, answers, on="Id")
```

```python
results = results.drop(columns = ["title"])
```
